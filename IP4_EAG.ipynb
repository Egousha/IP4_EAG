```python
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('jamesbond.csv')
```

```python
fig=plt.figure(figsize=(15,10))
heat_map=data[['US_Gross','US_Adj', 'World_Gross', 'World_Adj','Budget','Budget_Adj','Film_Length','Avg_User_IMDB','Avg_User_Rtn_Tom','Conquests','Martinis','BJB','Kills_Bond','Kills_Others']].corr(method='pearson')
cols=['US_Gross','US_Adj', 'World_Gross', 'World_Adj','Budget','Budget_Adj','Film_Length','Avg_User_IMDB','Avg_User_Rtn_Tom','Conquests','Martinis','BJB','Kills_Bond','Kills_Others']
axis=sns.heatmap(heat_map, annot=True, xticklabels=cols, yticklabels=cols )
```

```python
fig=plt.figure(figsize=(12,6)) 
plt.hist(data['World_Gross'],facecolor='lightgreen',
        edgecolor='blue',
        bins=10)
plt.title('Histogram of World Gross')
plt.show
```

```python
fig=plt.figure(figsize=(15,10))
b=sns.boxplot(y='Bond', x=data['World_Gross'].values, data=data)
```

```python
fig=plt.figure(figsize=(12,6)) 
plt.hist(data['Bond_Car_MFG'],facecolor='lightgreen',
        edgecolor='blue',
        bins=10)
plt.title('Histogram of the Bond Cars')
plt.show
```

```python
fig,axis1=plt.subplots()

fig=plt.figure(figsize=(15,10)) 

axis1.set_title('World_Gross vs Film Length')

axis1.set_xlabel('World_Gross')

axis1.tick_params(axis='x', labelcolor='blue',labelsize='large')

axis1.set_ylabel('Film Length', color='red', size='x-large')

axis1.tick_params(axis='y', labelcolor='red',labelsize='large')

axis1.scatter(data['World_Gross'],data['Film_Length'])

plt.show()
```

```python
fig,axis1=plt.subplots()

fig=plt.figure(figsize=(15,10)) 

axis1.set_title('Budget vs Imdb Rating')

axis1.set_xlabel('Budget')

axis1.tick_params(axis='x', labelcolor='blue',labelsize='large')

axis1.set_ylabel('IMDB Brating', color='red', size='x-large')

axis1.tick_params(axis='y', labelcolor='red',labelsize='large')

axis1.scatter(data['Budget'],data['Avg_User_IMDB'])

plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
ax1=fig.add_axes([0,0,1,1])
ax1.set_title('Budget through the Years')
ax1.plot(data['Year'],data['Budget'],color='green')
plt.show()
```

```python
fig=plt.figure(figsize=(12,6)) 
plt.hist(data['Bond'],facecolor='lightgreen',
        edgecolor='blue',
        bins=10)
plt.title('Histogram of the Bond Actors')
plt.show
```

```python

fig=plt.figure(figsize=(10,6))
ax1=fig.add_axes([0,0,1,1])
ax1.set_title('Bond Kills through the Years')
ax1.plot(data['Year'],data['Kills_Bond'],color='green')
plt.show()
```

```python
fig=plt.figure(figsize=(12,6)) 
plt.hist(data['Avg_User_IMDB'],facecolor='lightgreen',
        edgecolor='blue',
        bins=12)
plt.title('Histogram of the Imdb Ratings')
plt.show
```

```python
fig=plt.figure(figsize=(13,6))
plt.hist(data['Kills_Bond'],
        facecolor='lightgreen',
        edgecolor='blue',
        bins=10)
plt.hist(data['Kills_Others'],
        facecolor='blue',
        edgecolor='black',bins=10, alpha=.3)
plt.title('Comparative Histogram of Bond Kills vs Others Kills')
```

```python
fig=plt.figure(figsize=(13,6))
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.05,0.65,0.5,0.3])
ax1.set_title('World Gross')
ax2.set_title('Budget')
ax1.plot(data['Year'],data['World_Gross'],color='green')
ax2.plot(data['Year'],data['Budget'],color='blue')
plt.show()
```
